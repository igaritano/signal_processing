## Sine waves and FFT over them

```python
# Time vector generation

maxSec = 10 # Signal length in seconds

fs = 2000 # Sampling frequency = 1000Hz
ts = 1/fs # Sampling period = 1/fs = 0.001 sec

t = numpy.arange(0,maxSec,ts) # time vector from 0 to maxSec seconds

sampleNum = len(t) # Number of samples
print('Total number of samples = ', sampleNum)
```

### 1.- Simple sine wave

```python
# 1.-Sine wave

f = 1 # Signal frequency

signal = numpy.sin(2*numpy.pi*t*f)

fig = matplotlib.pyplot.figure(figsize=(18,14))
matplotlib.pyplot.subplot(411)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Signal of a sine, 1Hz frequency and amplitude=1')
matplotlib.pyplot.ylabel('Amplitude')
matplotlib.pyplot.plot(t,signal) # Plot

### diff ###

diffSignal = numpy.diff(signal)
diffSignal = numpy.concatenate((diffSignal, [0]))

matplotlib.pyplot.subplot(412)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Derivate of the signal of a sine, 1Hz frequency and amplitude=1')
matplotlib.pyplot.ylabel('Amplitude')
matplotlib.pyplot.plot(t,diffSignal) # Plot

### cumsum ###

cumsumSignal = numpy.cumsum(diffSignal)

matplotlib.pyplot.subplot(413)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Integral of the derivate of the signal of a sine, 1Hz frequency and amplitude=1')
matplotlib.pyplot.xlabel('Time in seconds (s)')
matplotlib.pyplot.ylabel('Amplitude')
matplotlib.pyplot.plot(t,cumsumSignal) # Plot

### FFT ### Taking into account only the first second

firstSecSampleNum = fs # Number of samples of the first second
halfFirstSecSampleNum = int(numpy.floor(firstSecSampleNum/2)) # Half of the number of samples of the first second

# Evaluate FFT considering only samples of the first second and normalize the output by dividing with the half number of samples
signalFFT = fft(signal[:firstSecSampleNum])/halfFirstSecSampleNum

# Evaluate the absolute value of the FFT and consider only half of the output since the rest is symetric
signalFFT = abs(signalFFT[:halfFirstSecSampleNum])**2

# Evaluate x axis considering the sampling frequency
axisX = numpy.arange(0,halfFirstSecSampleNum,(1/int(len(signalFFT)/halfFirstSecSampleNum)))
axisXlimit = 10

# Plot
matplotlib.pyplot.subplot(414)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('FFT of the signal of a sine, 1Hz frequency and amplitude=1')
matplotlib.pyplot.xlabel('Hz')
matplotlib.pyplot.ylabel('|Amplitude|^2')
matplotlib.pyplot.plot(axisX[:axisXlimit],signalFFT[:axisXlimit])

matplotlib.pyplot.tight_layout()
```

### 13.-Five sine waves with different amplitude and frequency

```python
# 13.-Five sine waves with different amplitude and frequency

phase1 = 0 # Signal phase for the 1st sine wave
phase2 = 0 # Signal phase for the 2nd sine wave
phase3 = 0 # Signal phase for the 3rd sine wave
phase4 = 0 # Signal phase for the 4th sine wave
phase5 = 0 # Signal phase for the 5th sine wave
amplitude1 = 4/(1*numpy.pi) # Signal amplitude for the 1st sine wave
amplitude2 = 4/(3*numpy.pi) # Signal amplitude for the 2nd sine wave
amplitude3 = 4/(5*numpy.pi) # Signal amplitude for the 3rd sine wave
amplitude4 = 4/(7*numpy.pi) # Signal amplitude for the 4th sine wave
amplitude5 = 4/(1*numpy.pi) # Signal amplitude for the 5th sine wave
f1 = 1 # Signal frequency for the 1st sine wave
f2 = 3 # Signal frequency for the 2nd sine wave
f3 = 5 # Signal frequency for the 3rd sine wave
f4 = 7 # Signal frequency for the 4th sine wave
f5 = 150 # Signal frequency for the 5th sine wave

signal = amplitude1*numpy.sin(2*numpy.pi*t*f1+phase1) \
+amplitude2*numpy.sin(2*numpy.pi*t*f2+phase2) \
+amplitude3*numpy.sin(2*numpy.pi*t*f3+phase3) \
+amplitude4*numpy.sin(2*numpy.pi*t*f4+phase4) \
+amplitude5*numpy.sin(2*numpy.pi*t*f5+phase5)

fig = matplotlib.pyplot.figure(figsize=(18,7))
matplotlib.pyplot.subplot(211)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Signal of a sine, 1Hz frequency and amplitude=0.5')
matplotlib.pyplot.ylabel('Amplitude')
matplotlib.pyplot.plot(t,signal) # Plot

### FFT ### Taking into account only the first second

firstSecSampleNum = fs # Number of samples of the first second
halfFirstSecSampleNum = int(numpy.floor(firstSecSampleNum/2)) # Half of the number of samples of the first second

# Evaluate FFT considering only samples of the first second and normalize the output by dividing with the half number of samples
signalFFT = fft(signal[:firstSecSampleNum])/halfFirstSecSampleNum

# Evaluate the absolute value of the FFT and consider only half of the output since the rest is symetric
signalFFT = abs(signalFFT[:halfFirstSecSampleNum])**2

# Evaluate x axis considering the sampling frequency
axisX = numpy.arange(0,halfFirstSecSampleNum,(1/int(len(signalFFT)/halfFirstSecSampleNum)))
axisXlimit = 160

# Plot
matplotlib.pyplot.subplot(212)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('FFT of sine signals')
matplotlib.pyplot.xlabel('Hz')
matplotlib.pyplot.ylabel('|Amplitude|^2')
matplotlib.pyplot.plot(axisX[:axisXlimit],signalFFT[:axisXlimit])

matplotlib.pyplot.tight_layout()
```

#### Design the filter to leave out 150Hz signal

```python
lowcut = 0.1 # Bandpass filter low frequency cutoff
highcut = 20.0 # Bandpass filter high frequency cutoff

fig = matplotlib.pyplot.figure(figsize=(18,7))
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Filter response')
matplotlib.pyplot.ylabel('Gain')
matplotlib.pyplot.xlabel('Hz')

for order in [2]:
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    w, h = freqz(b, a, worN=2000)
    matplotlib.pyplot.plot((fs * 0.5 / numpy.pi) * w, abs(h), label="order = %d" % order)
```

#### Filter the signal by lfilter and filtfilt

```python
# Filter the signal by lfilter method
lfilterFilteredSignal = lfilter(b, a, signal)

# Filter the signal by filtfilt method
filtfiltFilteredSignal = filtfilt(b, a, signal)

fig = matplotlib.pyplot.figure(figsize=(18,7))

matplotlib.pyplot.subplot(211)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Filtered signal by lfilter')
matplotlib.pyplot.ylabel('Amplitude')
matplotlib.pyplot.xlabel('Time (s)')
matplotlib.pyplot.plot(t,lfilterFilteredSignal) # plot

matplotlib.pyplot.subplot(212)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Filtered signal by filtfilt')
matplotlib.pyplot.ylabel('Amplitude')
matplotlib.pyplot.xlabel('Time (s)')
matplotlib.pyplot.plot(t,filtfiltFilteredSignal) # plot

matplotlib.pyplot.tight_layout()
```
