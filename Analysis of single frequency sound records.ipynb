# Analysis of single frequency sound records

#### Description

Frequency analysis of 100Hz, 250Hz and 3000Hz four different signals, some of them downloaded from mediacollege and others reproduced from youtube and recorded by a smartphone.

Source: http://www.mediacollege.com/audio/tone/download/

### Load necessary libraries

```python
### Libraries ###

# Matplotlib
import matplotlib
%matplotlib inline
matplotlib.rc("savefig", dpi=320)

# Numpy
import numpy

# Scipy fft
from scipy.fftpack import fft

# Scipy periodogram
from scipy.signal import periodogram

# Scipy io for wav files
from scipy.io import wavfile
```

### Load four signals and print their sampling frequency

```python
# Load signals

fs100Hz, data100Hz = wavfile.read('sources/100Hz_44100Hz_16bit_05sec.wav') # Read wav file composed of fs (sampling frequency) and data
print('Sampling frequency for 100Hz signal record = ', fs100Hz, 'Hz') # Sampling frequency

fs250Hz, data250Hz = wavfile.read('sources/250Hz_44100Hz_16bit_05sec.wav') # Read wav file composed of fs (sampling frequency) and data
print('Sampling frequency for 250Hz signal record = ', fs250Hz, 'Hz') # Sampling frequency

fs250HzSP, data250HzSP = wavfile.read('sources/250Hz_48000Hz_16bit_SP.wav') # Read wav file composed of fs (sampling frequency) and data
data250HzSP = data250HzSP.T[0]
print('Sampling frequency for 250Hz signal recorded by a smartphone = ', fs250HzSP, 'Hz') # Sampling frequency

fs3000HzSP, data3000HzSP = wavfile.read('sources/3000Hz_48000Hz_16bit_SP.wav') # Read wav file composed of fs (sampling frequency) and data
data3000HzSP = data3000HzSP.T[0]
print('Sampling frequency for 3000Hz signal recorded by a smartphone = ', fs3000HzSP, 'Hz') # Sampling frequency
```

### For each signal evaluate the sampling period and length or duration in seconds. Print the latest one.

```python
# Signal characteristics

ts100Hz = 1/fs100Hz # Sampling period
ts250Hz = 1/fs250Hz # Sampling period
ts250HzSP = 1/fs250HzSP # Sampling period
ts3000HzSP = 1/fs3000HzSP # Sampling period

d100Hz = data100Hz.shape[0]/fs100Hz # Signal duration in seconds
d250Hz = data250Hz.shape[0]/fs250Hz # Signal duration in seconds
d250HzSP = data250HzSP.shape[0]/fs250HzSP # Signal duration in seconds
d3000HzSP = data3000HzSP.shape[0]/fs3000HzSP # Signal duration in seconds

print('Signal length in seconds for 100Hz signal record = ', d100Hz, 'seconds')
print('Signal length in seconds for 250Hz signal record = ', d250Hz, 'seconds')
print('Signal length in seconds for 250Hz signal recorded by a smartphone = ', d250HzSP, 'seconds')
print('Signal length in seconds for 3000Hz signal recorded by a smartphone = ', d3000HzSP, 'seconds')
```

### For each signal select the 2nd second, normalize and plot first 1000 samples.

```python
# Select the 2nd second of the signal, normalize and plot it

samplesToPlot = 1000

fig = matplotlib.pyplot.figure(figsize=(18,14))

# 100Hz
matplotlib.pyplot.subplot(411)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Signal of a sine, 100Hz frequency')
matplotlib.pyplot.ylabel('Amplitude')

maxSignal100Hz = max(data100Hz)
minSignal100Hz = min(data100Hz)

if (maxSignal100Hz > abs(minSignal100Hz)):
    normalizationValue100Hz = maxSignal100Hz
else:
    normalizationValue100Hz = abs(minSignal100Hz)

signal100Hz = data100Hz/normalizationValue100Hz
signal100Hz2ndSec = signal100Hz[fs100Hz*1:fs100Hz*2]

matplotlib.pyplot.plot(signal100Hz2ndSec[:samplesToPlot]) # Plot

# 250Hz
matplotlib.pyplot.subplot(412)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Signal of a sine, 250Hz frequency')
matplotlib.pyplot.ylabel('Amplitude')

maxSignal250Hz = max(data250Hz)
minSignal250Hz = min(data250Hz)

if (maxSignal250Hz > abs(minSignal250Hz)):
    normalizationValue250Hz = maxSignal250Hz
else:
    normalizationValue250Hz = abs(minSignal250Hz)

signal250Hz = data250Hz/normalizationValue250Hz
signal250Hz2ndSec = signal250Hz[fs250Hz*1:fs250Hz*2]

matplotlib.pyplot.plot(signal250Hz2ndSec[:samplesToPlot]) # Plot

# 250HzSP
matplotlib.pyplot.subplot(413)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Signal of a sine, 250Hz frequency recorded by a smartphone')
matplotlib.pyplot.ylabel('Amplitude')

maxSignal250HzSP = max(data250HzSP)
minSignal250HzSP = min(data250HzSP)

if (maxSignal250HzSP > abs(minSignal250HzSP)):
    normalizationValue250HzSP = maxSignal250HzSP
else:
    normalizationValue250HzSP = abs(minSignal250HzSP)

signal250HzSP = data250HzSP/normalizationValue250HzSP
signal250HzSP2ndSec = signal250HzSP[fs250HzSP*1:fs250HzSP*2]

matplotlib.pyplot.plot(signal250HzSP2ndSec[:samplesToPlot]) # Plot

# 3000HzSP
matplotlib.pyplot.subplot(414)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Signal of a sine, 3000Hz frequency recorded by a smartphone')
matplotlib.pyplot.ylabel('Amplitude')

maxSignal3000HzSP = max(data3000HzSP)
minSignal3000HzSP = min(data3000HzSP)

if (maxSignal3000HzSP > abs(minSignal3000HzSP)):
    normalizationValue3000HzSP = maxSignal3000HzSP
else:
    normalizationValue3000HzSP = abs(minSignal3000HzSP)

signal3000HzSP = data3000HzSP/normalizationValue3000HzSP
signal3000HzSP2ndSec = signal3000HzSP[fs3000HzSP*1:fs3000HzSP*2]

matplotlib.pyplot.plot(signal3000HzSP2ndSec[:samplesToPlot]) # Plot
```

### FFT analysis of each signal

```python
### FFT ###

fig = matplotlib.pyplot.figure(figsize=(18,14))

# 100Hz
matplotlib.pyplot.subplot(411)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Frequency of a sine, 100Hz frequency')
matplotlib.pyplot.ylabel('|Amplitude|')

halfEachSecSampleNum100Hz = int(numpy.floor(fs100Hz/2)) # Half of the number of samples of each second

# Evaluate FFT considering only samples of the 2nd second and normalize the output by dividing with the half number of samples
signalFFT100Hz = fft(signal100Hz2ndSec)/halfEachSecSampleNum100Hz

# Evaluate the absolute value of the FFT and consider only half of the output since the rest is symetric (Nyquist)
signalFFT100Hz = abs(signalFFT100Hz[:halfEachSecSampleNum100Hz])

# Evaluate x axis considering the sampling frequency
axisX100Hz = numpy.arange(0,halfEachSecSampleNum100Hz,(1/int(len(signalFFT100Hz)/halfEachSecSampleNum100Hz)))
minAxisXlimit100Hz = 90
maxAxisXlimit100Hz = 110

# Plot
matplotlib.pyplot.plot(axisX100Hz[minAxisXlimit100Hz:maxAxisXlimit100Hz],signalFFT100Hz[minAxisXlimit100Hz:maxAxisXlimit100Hz])

# 250Hz
matplotlib.pyplot.subplot(412)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Frequency of a sine, 250Hz frequency')
matplotlib.pyplot.ylabel('|Amplitude|')

halfEachSecSampleNum250Hz = int(numpy.floor(fs250Hz/2)) # Half of the number of samples of each second

# Evaluate FFT considering only samples of the 2nd second and normalize the output by dividing with the half number of samples
signalFFT250Hz = fft(signal250Hz2ndSec)/halfEachSecSampleNum250Hz

# Evaluate the absolute value of the FFT and consider only half of the output since the rest is symetric (Nyquist)
signalFFT250Hz = abs(signalFFT250Hz[:halfEachSecSampleNum250Hz])

# Evaluate x axis considering the sampling frequency
axisX250Hz = numpy.arange(0,halfEachSecSampleNum250Hz,(1/int(len(signalFFT250Hz)/halfEachSecSampleNum250Hz)))
minAxisXlimit250Hz = 240
maxAxisXlimit250Hz = 260

# Plot
matplotlib.pyplot.plot(axisX250Hz[minAxisXlimit250Hz:maxAxisXlimit250Hz],signalFFT250Hz[minAxisXlimit250Hz:maxAxisXlimit250Hz])

# 250HzSP
matplotlib.pyplot.subplot(413)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Frequency of a sine, 250Hz frequency recorded by a smartphone')
matplotlib.pyplot.ylabel('|Amplitude|')

halfEachSecSampleNum250HzSP = int(numpy.floor(fs250HzSP/2)) # Half of the number of samples of each second

# Evaluate FFT considering only samples of the 2nd second and normalize the output by dividing with the half number of samples
signalFFT250HzSP = fft(signal250HzSP2ndSec)/halfEachSecSampleNum250HzSP

# Evaluate the absolute value of the FFT and consider only half of the output since the rest is symetric (Nyquist)
signalFFT250HzSP = abs(signalFFT250HzSP[:halfEachSecSampleNum250HzSP])

# Evaluate x axis considering the sampling frequency
axisX250HzSP = numpy.arange(0,halfEachSecSampleNum250HzSP,(1/int(len(signalFFT250HzSP)/halfEachSecSampleNum250HzSP)))
minAxisXlimit250HzSP = 240
maxAxisXlimit250HzSP = 260

# Plot
matplotlib.pyplot.plot(axisX250HzSP[minAxisXlimit250HzSP:maxAxisXlimit250HzSP],signalFFT250HzSP[minAxisXlimit250HzSP:maxAxisXlimit250HzSP])

# 3000HzSP
matplotlib.pyplot.subplot(414)
matplotlib.pyplot.grid(linestyle=':', linewidth=0.5)
matplotlib.pyplot.title('Frequency of a sine, 3000Hz frequency recorded by a smartphone')
matplotlib.pyplot.ylabel('|Amplitude|')

halfEachSecSampleNum3000HzSP = int(numpy.floor(fs3000HzSP/2)) # Half of the number of samples of each second

# Evaluate FFT considering only samples of the 2nd second and normalize the output by dividing with the half number of samples
signalFFT3000HzSP = fft(signal3000HzSP2ndSec)/halfEachSecSampleNum3000HzSP

# Evaluate the absolute value of the FFT and consider only half of the output since the rest is symetric (Nyquist)
signalFFT3000HzSP = abs(signalFFT3000HzSP[:halfEachSecSampleNum3000HzSP])

# Evaluate x axis considering the sampling frequency
axisX3000HzSP = numpy.arange(0,halfEachSecSampleNum3000HzSP,(1/int(len(signalFFT3000HzSP)/halfEachSecSampleNum3000HzSP)))
minAxisXlimit3000HzSP = 3140
maxAxisXlimit3000HzSP = 3160

# Plot
matplotlib.pyplot.plot(axisX3000HzSP[minAxisXlimit3000HzSP:maxAxisXlimit3000HzSP],signalFFT3000HzSP[minAxisXlimit3000HzSP:maxAxisXlimit3000HzSP])
```

As we see, 3000Hz signal is displaced in frequency up to 3150Hz.
